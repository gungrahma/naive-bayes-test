```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.naive_bayes import CategoricalNB
data_pasien = pd.DataFrame({
    'ID' : [1, 2, 3, 4, 5],
    'Gejala' : ['Sakit Kuning', 'Pusing', 'Sakit Kuning', 'Sesak Nafas', 'Pusing'],
    'Tindakan_Awal' : ['Obat', 'Istirahat', 'Infus', 'Oksigen', 'Obat'],
    'Masuk_UGD' : [1, 0, 1, 1, 0]
})

print("Data pasien berhasil dimuat (DUMMY DATA)")
display(data_pasien)
```

```python
X_kategori = data_pasien[['Gejala', 'Tindakan_Awal']]
Y_target = data_pasien['Masuk_UGD']

encoder = OrdinalEncoder()
X_angka = encoder.fit_transform(X_kategori)

model_nb_kategori = CategoricalNB()
model_nb_kategori.fit(X_angka, Y_target)
print("Data berhasil dilatih oleh Naive Bayes!")
```

```python
pasien_baru = pd.DataFrame({
    'Gejala': ['Sakit Kuning'],
    'Tindakan_Awal': ['Infus']
})

pasien_baru_angka = encoder.transform(pasien_baru)

prediksi = model_nb_kategori.predict(pasien_baru_angka)
probabilitas = model_nb_kategori.predict_proba(pasien_baru_angka)

print("\nHasil Prediksi Naive Bayes")
persentase_yakin = probabilitas[0][1] * 100 if prediksi[0] == 1 else probabilitas [0][0] * 100
status = "Masuk UGD (1)" if prediksi[0] == 1 else "Boleh pulang"

print(f"Kondisi: Gejala {pasien_baru['Gejala'][0]}, Diberi {pasien_baru['Tindakan_Awal'][0]}")
print(f"Keputusan: {status} (Yakin: {persentase_yakin: .1f}%)")
```
