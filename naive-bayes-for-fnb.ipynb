```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

ulasan = np.array([
    "ayamnya krispi bumbunya meresap",      
    "makanan basi bau asem",                
    "pengiriman cepat dan kurir ramah",     
    "lama banget nunggu sejam perut lapar", 
    "porsi besar harga sangat murah",       
    "kuah tumpah kemasan hancur lebur",     
    "enak parah bakal order lagi besok",    
    "rasanya hambar dagingnya keras"
])

label = np.array([1, 0, 1, 0, 1, 0, 1, 0])
print("Data review restaurant X berhasil dimuat!")
```

```python
vectorizer = CountVectorizer()
X_angka = vectorizer.fit_transform(ulasan)

model_nb = MultinomialNB()

model_nb.fit(X_angka, label)
print("Model Naive Bayes berhasil dilatih!")
```

```python
ulasan_baru = [
    "ayamnya krispi bumbunya meresap porsi besar harga sangat murah pengiriman cepat enak parah"
]

X_baru = vectorizer.transform(ulasan_baru)

prediksi = model_nb.predict(X_baru)
probabilitas = model_nb.predict_proba(X_baru)

print("Hasil deteksi sentimen otomatis")
for i in range(len(ulasan_baru)):
    teks = ulasan_baru[i]
    persen_yakin = probabilitas[i][1] * 100 if prediksi[i] == 1 else probabilitas[i][0] * 100
    status_teks = "Puas (1)" if prediksi[i] == 1 else "Kecewa (0)"

    print(f"Ulasan: '{teks}'")
    print(f"Sentimen: {status_teks} (Yakin {persen_yakin: .4f}%)\n")
```
